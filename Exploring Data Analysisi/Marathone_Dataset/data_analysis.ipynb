```python
# combine all filters together
```

```python
dataframe2 = dataframe[(dataframe['Event distance/length'].isin(['50km','50mi'])) & (dataframe['Year of event'] == 2020) & (dataframe['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == "USA")]
```

```python
dataframe2.head(10)
```

```python
dataframe2.shape
```

```python
# Remove USA from the event name 
dataframe2['Event name'] = dataframe2['Event name'].str.split('(').str.get(0)
```

```python
dataframe2.head(10)
```

```python
# clean up athelet age
dataframe2['athelete_age'] = 2020 - dataframe2['Athlete year of birth']
```

```python
# remove h from the athelete performance
dataframe2['Athlete performance']  = dataframe2['Athlete performance'].str.split(' ').str.get(0)
```

```python
dataframe2.head(10)
```

```python
# drop some columns: Athelete club, Athelete country, Athelete year of birth, athelete age category
dataframe2 = dataframe2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
dataframe2.head(10)
```

```python
# clean up null values
dataframe2.isna().sum()
# This will tell us where the nulls are 
```

```python
dataframe2[dataframe2['athelete_age'].isna() == 1]
```

```python
dataframe2 = dataframe2.dropna()
dataframe2.shape
```

```python
# check for dublicate values
dataframe2[dataframe2.duplicated() == True]
# nth here is dublicated
```

```python
# reset index
dataframe2.reset_index(drop= True)
# this way we dont get the wierd index columns
```

```python
# fix types
```

```python
dataframe2.dtypes
```

```python
dataframe2.loc[:,'athlete_age'] = dataframe2['athelete_age'].astype(int)
```

```python
dataframe2['Athlete average speed'] = dataframe2['Athlete average speed'].astype(float)
```

```python
dataframe2.dtypes
```

```python
dataframe2.head(5)
```

```python
# Rename columns
dataframe2 = dataframe2.rename(columns = {
    'Year of event':'year',
    'Event dates':'race_day',
    'Event name':'race_name',
    'Event distance/length':'race_length',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance':'athlete_performance',
    'Athlete gender':'athlete_gender',
    'Athlete average speed':'athlete_average_speed',	
    'Athlete ID':'athlete_id',
})
```

```python
dataframe2.head(2)
```

```python
# reorder columns
```

```python
dataframe3 = dataframe2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
dataframe3.head(2)
```

```python
dataframe3[dataframe3['athlete_id'] == 222509]
```

```python
# Histogram using seaborn
# sns.histplot(dataframe3,x='race_length',hue='athlete_gender')

# histogram for male and female in the 50km and 50mi race

male_data = dataframe3[dataframe3['athlete_gender'] == 'M']['race_length']
female_data = dataframe3[dataframe3['athlete_gender'] == 'F']['race_length']
print(male_data,female_data)
plt.hist(male_data, bins=20, alpha=0.5, label='Male', color='blue')
plt.hist(female_data, bins=20, alpha=0.5, label='Female', color='pink')
plt.xlabel('Race length')
plt.ylabel('Race Frequency')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Using seaborn
sns.displot(dataframe3[dataframe3['race_length'] == '50mi']['athlete_average_speed'])
```

```python
data_displot = dataframe3[dataframe3['race_length'] == '50mi']['athlete_average_speed']

plt.hist(data_displot,bins=20,density=True,alpha=0.6,color='blue')
sns.kdeplot(data_displot,color='red')

plt.xlabel('Athlete Average Speed')
plt.ylabel('Density')
plt.title('Distribution of Athlete Average Speed for 50mi Race')
plt.show()
# data_displot
```

```python
sns.violinplot(data = dataframe3, x='race_length', y = 'athlete_average_speed', hue='athlete_gender')
```

```python
sns.lmplot(data = dataframe3,x= 'athlete_age', y = 'athlete_average_speed',hue = 'athlete_gender')
```

```python
# Difference in speed for the 50k,50mi male and female 
dataframe3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age group are the best in the 50km Race (20 + races min)
dataframe3[dataframe3['race_length'] == '50mi']\
.groupby('athlete_age')['athlete_average_speed']\
.agg(['mean','count'])\
.sort_values('mean',ascending=False)\
.query('count>19')\
.head(15)
```
