#  Setup loss function and optimizer

Which loss function or optimizer shouhld you use ?

Again... this is problem specific.

For example for regression you might want MAE or MSE (mean absolute error or mean squared error).

For classification you might want binary cross entropy or categorical cross entropy (cross entropy)

And for optimizers, two of the most common and useful are SGD and Adam, however PyTorch has many buit-in options.

* For the loss function we're going to use `torch.nn.BCEWithLogitsLoss()`
* For different optimiers see `torch.optim`

**Note** the nn.BCELoss function requires a sigmoid function to be written before it in order for it to work like a BCEWithLogitsLoss() but it is not numerically stable than `nn.BCEWithLogitsLoss()`

```python
# Steup the loss fucntion 
# loss_fn = nn.BCELoss() # requires inputs to have gone through the sigmoid activation function prior to the input tp BCELoss
loss_fn =nn.BCEWithLogitsLoss()  # This has the sigmoid activation function buit in 
optmiizer = torch.optim.SGD(params=model_0.parameters(),
                           lr=0.1
                           )
```

```python
model_0.state_dict()
```

```python
# Calcualte accuracy 
# formula for accuracy = True Positive / (True Positive + True Negative) * 100
# accuracy = out of the 100 examples what percentage does our model gets right?
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true,y_pred).sum().item()
    acc = (correct / len(y_pred)) * 100
    return acc
```

### Training a model

To train our model,we need a traiing loop with the following steps

1. Forward pass
2. calcualte loss
3. optimise zero grad
4. loss backward (back propagation)
5. optimizer (gradient decent)

### Going from raw logits -> prediction probabilities -> prediction labels

Our model output are going to be raw **LOGITS**

We can convert these **logits** into prediction probabilities by passing them to some kind of activation function (e.g. sigmoid for bianry classification  and softmax for multiclass classificaiton)

Then we can convert out models predcition proababilites to **prediction labels** by either roundinf them or taking the `argmax()`

## Building a training and testing loop
the `BCEWithLogitsLoss` expects logits with inputs so while calculating the loss we need to pass the logits

```python
torch.manual_seed(42)
torch.cuda.manual_seed(42)

# set the number of epochs
epochs = 1000

# Put the data to the taret device
X_train, y_train = X_train.to(device), y_train.to(device)
X_test, y_test = X_test.to(device), y_test.to(device)


#Building training and evaluation loop
for epoch in range(epochs):
    ### traiing
    model_0.train()

    # forward pass
    y_logits = model_0(X_train).squeeze()
    y_pred = torch.round(torch.sigmoid(y_logits))

    # Calcaulte the loss and accuracy
    # loss = loss_fn(torch.sigmoid(y_logits), # nn.BCELoss expects prediction probabliites as inputs
    #                y_train
    #                )
    loss= loss_fn(y_logits, # nn.BCEWithLogistsLoss expects raw logits as inputs
                  y_train)
    acc = accuracy_fn(y_true=y_train,
                     y_pred = y_pred)

    # Optimizer zerp grad 
    optmiizer.zero_grad()

    # loss backward (Back propagation)
    loss.backward()

    # optmizer step (gradient descent)
    optmiizer.step()

    ## Testing
    model_0.eval()
    with torch.inference_mode():
        # 1. forward pass
        test_logits = model_0(X_test).squeeze()
        test_pred = torch.round(torch.sigmoid(test_logits))
        #2. Calcualte the test loss
        test_loss = loss_fn(test_logits,
                           y_test
                           )
        test_acc = accuracy_fn(y_true=y_test,y_pred=test_pred)

    # PRint out 
    if epoch % 10 == 0:
        print(f"Epoch: {epoch} | Loss: {loss:.4f} | Acc: {acc:.4f} | Test Loss : {test_loss:.4f} | Test acc: {test_acc:.4f}%")
        
    
```

```python
class CircleModelV1(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(in_features=2,out_features=10)
        self.layer_2 = nn.Linear(in_features= 10, out_features=10)
        self.layer_3 = nn.Linear(in_features=10,out_features=1)

    def forward(self,x):
        return self.layer_3(self.layer_2(self.layer_1(x)))



model_1 = CircleModelV1().to(device)
model_1
```

```python
# Create a loss function
lossfn = nn.BCEWithLogitsLoss()
# Create an optimizer
optmiizer = torch.optim.SGD(params=model_1.parameters(),lr=0.01)
model_1.state_dict()
```

```python
# Write a training and evalation loop fro model_1
torch.manual_seed(42)
torch.cuda.manual_seed(42)

epochs = 1000

X_test,y_test = X_test.to(device),y_test.to(device)
X_train,y_train = X_train.to(device),y_train.to(device)

# Building training and evaluation loop
for epoch in range(epochs):
    # 1.Training
    model_1.train()
    # 2. Forward pass 
    y_logits = model_1(X_train).squeeze()
    y_pred = torch.round(torch.sigmoid(y_logits)) 
    # 3. Calculate the loss
    loss = loss_fn(y_logits,y_train)
    acc = accuracy_fn(y_true=y_train,y_pred= y_pred)

    # 4. optimize zero grad
    optmiizer.zero_grad()

    #5. back propagation
    loss.backward()


    # 6. gradient descent
    optmiizer.step()

    model_1.eval()

    with torch.inference_mode():
        # forward pass
        test_logits = model_1(X_test).squeeze()
        test_pred = torch.round(torch.sigmoid(test_logits))

        # test loss
        test_loss = loss_fn(test_logits,y_test)
        test_acc = accuracy_fn(y_true=y_test,y_pred=test_pred)


    if epoch % 100 == 0:
        print(f"Epoch: {epoch} | Loss: {loss:4f} , Acc: {acc : 4f} | Test Loss: {test_loss:4f} | Test Acc: {test_acc:4f}% ")
    
```

## Building a model with mom-linearity
* linear = straight line
* Non-linear = non-straight line

Artificial neural networks are a combination of linear and non linear functions which are potentially able to find patterns in genreating data 

```python
class CircleModelV2(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(in_features=2,out_features=20)
        self.layer2 = nn.Linear(in_features=20, out_features=20)
        self.layer3 = nn.Linear(in_features=20,out_features = 1)
        self.relu = nn.ReLU()

    def forward(self,x):
        return self.layer3(self.relu(self.layer2(self.relu(self.layer1(x)))))

model_3 = CircleModelV2().to(device)
model_3
```

```python
# creating a loss function 
loss = nn.BCEWithLogitsLoss()

optimizer = torch.optim.SGD(model_3.parameters(),lr=0.1)
```

```python
# Random seed
torch.manual_seed(42)
torch.cuda.manual_seed(42)


X_train,y_train = X_train.to(device), y_train.to(device)
X_test,y_test = X_test.to(device), y_test.to(device)

epochs = 5000

# training
for epoch in range(epochs):
    model_3.train()

    # forward pass
    y_logits = model_3(X_train).squeeze()
    y_pred = torch.round(torch.sigmoid(y_logits))


    # loss calculation and accuracy calculation 
    loss = loss_fn(y_logits,y_train)
    acc = accuracy_fn(y_true=y_train,
                     y_pred=y_pred)


    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    #Testing
    model_3.eval()
    with torch.inference_mode():
        test_logits = model_3(X_test).squeeze()
        test_pred = torch.round(torch.sigmoid(test_logits))

        test_loss = loss_fn(test_logits,y_test)

        test_acc = accuracy_fn(y_true=y_test,y_pred=test_pred)

    if epoch % 500 == 0:
        print(f"Epoch: {epoch} | Loss: {loss:.4f} | Acc: {acc:.4f}% | Test Loss : {test_loss:.4f} | Test acc: {test_acc:.4f}%")
        
    
```

### Buliiding a multiclass classification model in pytorch

```python
# Create device agnostoc code
device = "cuda" if torch.cuda.is_available() else "device"
device
```

```python
# Build a multiclass classificaiton model
class BlobModel(nn.Module):
    def __init__(self,input_features, out_features, hidden_units = 8):
        super().__init__()
        self.linear_layer_stack = nn.Sequential(
            nn.Linear(in_features=input_features,out_features=hidden_units),
            # nn.ReLU(),
            nn.Linear(in_features=hidden_units,out_features=hidden_units),
            # nn.ReLU(),
            nn.Linear(in_features=hidden_units,out_features=out_features)
        )

    def forward(self,x):
        return self.linear_layer_stack(x)

# Creating an instance of blobModel and send it to target device
model_4 = BlobModel(input_features=2,
                   out_features=4,
                   hidden_units=8).to(device)
model_4
```

```python
# loss function and optimizer for multiclass classfication model
loss_fn =nn.CrossEntropyLoss()

optimizer = torch.optim.SGD(params = model_4.parameters(),
                            lr=0.1)
```

### Create a training loop and testing loop for a multi calss PyTorch model

```python
# Fit the multiclass model to the data
torch.manual_seed(42)
torch.cuda.manual_seed(42)

epochs = 100

# Put the data to the target device
X_blob_train,y_blob_train = X_blob_train.to(device), y_blob_train.to(device)
X_blob_test,y_blob_test = X_blob_test.to(device), y_blob_test.to(device)

for epoch in range(epochs):
    model_4.train()

    y_logits = model_4(X_blob_train) #These are the logits
    y_pred = torch.softmax(y_logits,dim=1).argmax(dim=1) # and we are going to prediction probabilities

    loss = loss_fn(y_logits,y_blob_train)
    acc = accuracy_fn(y_true=y_blob_train,
                     y_pred = y_pred)


    optimizer.zero_grad()
    loss.backward()
    optimizer.step() # optmizer will update the model to better represent the patterns in out training data

    model_4.eval()
    with torch.inference_mode():
        test_logits = model_4(X_blob_test)
        test_pred = torch.softmax(test_logits,dim=1).argmax(dim=1)

        test_loss = loss_fn(test_logits,y_blob_test)

        test_acc = accuracy_fn(y_true=y_blob_test,
                              y_pred = test_pred)

        # print 

    if epochs % 2 == 0:
        print(f"Epochs:{epoch} | Loss: {loss:.4f} , Acc: {acc:.2f}% | Test loss: {test_loss:.4f} Test Acc: {test_acc:4f}%")

    
```
